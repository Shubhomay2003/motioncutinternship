```python
def add_expense(expenses):
    try:
        amount = float(input("Enter the amount spent: "))
        description = input("Enter a brief description: ")
        category = input("Enter the category (e.g., food, transportation, entertainment): ")

        expenses.append({
            'amount': amount,
            'description': description,
            'category': category
        })

        print("Expense added successfully!")
    except ValueError:
        print("Invalid input. Please enter a numeric value for the amount.")

    return expenses
```

```python
def show_summary(expenses):
    total_expense = sum(expense['amount'] for expense in expenses)
    print(f"\nTotal Expenses: {total_expense}")

    categories = {}
    for expense in expenses:
        category = expense['category']
        amount = expense['amount']
        if category in categories:
            categories[category] += amount
        else:
            categories[category] = amount
    

    print("\nCategory-wise expenses:")
    for category, total in categories.items():
        print(f"{category}: {total}")
    import matplotlib.pyplot as plt

    print("Percentage distribution of your overall expenses:-")
    # Prepare data for plotting
    categories = [expense['category'] for expense in expenses]
    amounts = [expense['amount'] for expense in expenses]
    
    # Create the pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(amounts, labels=categories, autopct='%1.1f%%')

    # Adding a title
    plt.title('Expenditure Distribution by Category')

    # Show the plot
    plt.show()
```

```python
def menu():
    print("\n---------------------------------------------------------------------------")
    print("                     Welcome to Expense Tracker System ")
    print("---------------------------------------------------------------------------")
    print("    1. Add Expense")
    print("    2. View Summary")
    print("    3. Exit")
    print("---------------------------------------------------------------------------")
```

```python
expenses = []

while True:
    menu()
    choice = input("Choose an option: ")

    if choice == '1':
        expenses = add_expense(expenses)
    elif choice == '2':
        show_summary(expenses)
    elif choice == '3':
        print("Exiting the program. Thanks for using us!")
        break
    else:
        print("Invalid choice. Please try again.")
```
